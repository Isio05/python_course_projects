```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = sm.datasets.macrodata.load_pandas().data
index = pd.Index(sm.tsa.datetools.dates_from_range('1959Q1','2009Q3'))
data.index = index
```

```python
gdp_cycle, gdp_trend = sm.tsa.filters.hpfilter(data['realgdp'])
data['trend'] = gdp_trend
data[['realgdp','trend']]['2000-01-01':].plot(figsize=(12,4))
```

```python
passengers = pd.read_csv('airline_passengers.csv',index_col='Month')
passengers.head()
```

```python
passengers.dropna(inplace=True)
passengers.index = pd.to_datetime(passengers.index)
passengers.head()
```

```python
passengers['20 SMA'] = passengers['Thousands of Passengers'].rolling(window=7).mean()
passengers[['20 SMA','Thousands of Passengers']].plot(figsize=(12,4))
```

```python
passengers['20 EWMA'] = passengers['Thousands of Passengers'].ewm(span=20).mean()
passengers['20 EWMA'].plot()
```

```python
#Seasonal decomposition
result = sm.tsa.seasonal_decompose(passengers['Thousands of Passengers'])
result.trend.plot()
```
